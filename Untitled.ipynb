```python
def load_file(filename):
    with open(filename,encoding='utf-8') as f:
        data = f.read()
    return data
        
data = load_file("dataset/english_speech_2.txt")
text = data
print(text[:100])
```

```python
#text = "hello helli hello"
```

```python
k = 4

T = {}

for i in range(len(text)-k):
    X = text[i:i+k]
    Y = text[i+k]
    
    #Our task is to anaylyze probabilites
    
    #"Hell" is not present
    if T.get(X) is None:
        T[X] = {}
        T[X][Y] = 1
    else:
        # "Hell" is present but 'Hell''i' is not present
        if T[X].get(Y) is None:
            T[X][Y] =1
        else:
            T[X][Y] +=1
            
#Iterate over keys in T and Sum
for kx in T.keys():
    s = 0
    for v in T[kx].values():
        s += v
        #divide every term in the row with sum
    for ky in T[kx]:
        T[kx][ky] /= s
```
